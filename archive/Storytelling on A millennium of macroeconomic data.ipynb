```python
df = pd.read_csv('millenniumofdata_v3_headlines.csv')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.columns
```

```python
df = df.iloc[1: , :]
```

```python
for column in df.columns:
    df[column] = pd.to_numeric(df[column])

```

```python
columns=df.columns
columns=columns[1:]
```

```python
df[columns].describe()
```

As we can see here, some columns can be best described by a line plot, others by a scatter plot, and some by a waterfall plot. So let's explore a lottle bit more. For example, real GDP is best describe by a line since we are trying to track trends through time

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111)
sns.lineplot(data=df,x='Description',y='Real GDP of England at market prices')
plt.xlabel('Year')
plt.show()
```

Furthermore, we can plot employment vs year, and add the unemployment rate by setting up a bubble graph

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111)
sns.scatterplot(data=df,x='Description',y='Employment',size='Unemployment rate')
plt.xlabel('Year')
plt.show()
```

We can also use an area chart to compare export and import over time.

```python
df1=df[['Description','Export prices', 'Import prices']]
```

```python
fig = plt.figure(figsize=(18,14))
ax = fig.add_subplot(111)
plt.stackplot(df1['Description'], df1['Export prices'], df1['Import prices'],
              labels=['Export', 'Import'])
plt.show()
```
