```python
SELECT DISTINCT concat(m.firstname,' ',m.surname,' ',f.name) 
FROM Facilities AS f 
INNER JOIN Bookings AS b 
USING (facid) 
INNER JOIN Members AS m 
USING (memid) 
Where f.facid IN (0,1);
```

```python
SELECT concat(f.name,' ',m.firstname,' ',m.surname) AS facility_and_member, IF (memid=0,guestcost*slots,membercost*slots) AS Cost 
FROM Facilities AS f 
INNER JOIN Bookings AS b 
USING (facid) 
INNER JOIN Members AS m 
USING (memid) 
Where starttime LIKE '2012-09-14%' 
AND ((memid=0 AND slots*guestcost>30) OR (memid>0 AND slots*membercost>30)) 
ORDER BY cost DESC;
```

```python
SELECT tt.facility_and_member, tt.Cost 
FROM(
SELECT concat(f.name,' ',m.firstname,' ',m.surname) AS facility_and_member, IF (memid=0,guestcost*slots,membercost*slots) AS Cost 
FROM Facilities AS f 
INNER JOIN Bookings AS b 
USING (facid) 
INNER JOIN Members AS m 
USING (memid) 
Where starttime LIKE '2012-09-14%' ) as tt
WHERE tt.Cost > 30
ORDER BY tt.cost DESC;
```

```python
from sqlalchemy import create_engine
import pandas  as pd
```

```python
engine=create_engine("sqlite:///sqlite_db_pythonsqlite.db")
```

```python
query="SELECT f.name, SUM(IIF(memid=0,guestcost*slots,membercost*slots)) AS revenue FROM 'Facilities' AS f INNER JOIN 'Bookings' AS b USING (facid) GROUP BY f.name HAVING revenue < 1000 ORDER BY revenue "
df=pd.read_sql_query(query,engine)
df
```

```python
query="SELECT DISTINCT  m.memid, m.firstname AS MFirst, m.surname MSur,r.firstname RFirst, r.surname AS RSur FROM 'Members' AS m LEFT JOIN 'Members' AS r on m.memid = r.recommendedby WHERE m.memid>0 "
df=pd.read_sql_query(query,engine)
df
```

```python
query="SELECT f.name, COUNT(bookid) AS usage FROM 'Facilities' AS f INNER JOIN 'Bookings' AS b USING (facid) GROUP BY f.name HAVING memid>0"
df=pd.read_sql_query(query,engine)
df
```

```python
query="SELECT strftime('%m', starttime) as Month, COUNT(bookid) AS usage FROM 'Facilities' AS f INNER JOIN 'Bookings' AS b USING (facid) GROUP BY Month HAVING memid>0"
df=pd.read_sql_query(query,engine)
df
```
